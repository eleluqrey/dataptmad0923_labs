## Challenge 1
#### 1.- Define a function that simulates rolling a dice 10 times. Save the information in a dataframe.
**Hint**: you can use the *choices* function from module *random* to help you with the simulation.

```python
# your code here
```

```python
def simulate_dice_rolls(num_rolls=10, num_faces=6):
    # Simulación de los 10 lanzamientos aleatorios con las 6 caras de un dado usando la función choices
    dice_rolls = choices(range(1, num_faces + 1), k=num_rolls)

    # DataFrame con una columna de las tiradas y otra de los resultados al lanzar el dado
    df = pd.DataFrame({'Roll': range(1, num_rolls + 1), 'Result': dice_rolls})

    return df


dice_rolls_df = simulate_dice_rolls()
print(dice_rolls_df)
```

#### 2.- Plot the results sorted by value.

```python
# your code here
```

```python
# Ordeno con sort_values

dice_rolls_df_sorted = dice_rolls_df.sort_values(by='Result')
```

```python
print(dice_rolls_df_sorted)
```

```python
# Plot con bar charts

plt.bar(dice_rolls_df_sorted['Roll'], dice_rolls_df_sorted['Result'])
plt.xlabel('Lanzamiento')
plt.ylabel('Resultado')
plt.title('Resultados de lanzamiento de dado ordenados')
plt.show()
```

#### 3.- Calculate the frequency distribution and plot it. What is the relation between this plot and the plot above? Describe it with words.

```python
# your code here
```

```python
# Distribución de frecuencias con value_counts

frequency_distribution = dice_rolls_df['Result'].value_counts().sort_index()
type(frequency_distribution)

# value_counts() cuenta la frecuencia de cada valor en la columna Result, y sort_index() se utiliza para ordenar los valores del dado en orden ascendente
```

```python
# Plot con bar charts

plt.bar(frequency_distribution.index , frequency_distribution.values)
plt.xlabel('Resultado del dado')
plt.ylabel('Frecuencia')
plt.title('Distribución de frecuencias de resultados de dados')
plt.show()
```

```python
# Nota

# Eje x - frequency_distribution.index: Esto representa los valores únicos del dado, que son los índices de la Serie
# frequency_distribution. Son los posibles resultados del dado (1, 2, 3, 4, 5, 6).

# Eje y - frequency_distribution.values: Esto representa las frecuencias asociadas con cada valor del dado. Son los conteos
# de cuántas veces se obtuvo cada valor en los lanzamientos.
```
